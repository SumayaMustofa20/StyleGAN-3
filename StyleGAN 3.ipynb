```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
from pathlib import Path
from torchvision import transforms
from torchvision.models import inception_v3
from scipy.linalg import sqrtm
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
import json
from tqdm import tqdm
import warnings
import gc
warnings.filterwarnings('ignore')

# Memory management
def clear_memory():
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

class ComprehensiveMetrics:
    def __init__(self, device='cuda'):
        self.device = device
        # Load lighter inception model for FID
        self.inception = inception_v3(pretrained=True, transform_input=False).to(device)
        self.inception.eval()

    def get_inception_features(self, images):
        """Get inception features for FID calculation"""
        with torch.no_grad():
            # Resize images to 299x299 for Inception
            images_resized = F.interpolate(images, size=(299, 299), mode='bilinear', align_corners=False)
            features = self.inception(images_resized)
            return features.cpu().numpy()

    def calculate_fid(self, real_features, fake_features):
        """Calculate FrÃ©chet Inception Distance"""
        # Calculate mean and covariance
        mu_real, sigma_real = np.mean(real_features, axis=0), np.cov(real_features, rowvar=False)
        mu_fake, sigma_fake = np.mean(fake_features, axis=0), np.cov(fake_features, rowvar=False)

        # Calculate FID
        diff = mu_real - mu_fake
        covmean = sqrtm(sigma_real.dot(sigma_fake))

        if np.iscomplexobj(covmean):
            covmean = covmean.real

        fid = diff.dot(diff) + np.trace(sigma_real) + np.trace(sigma_fake) - 2 * np.trace(covmean)
        return fid

    def calculate_ssim_psnr_per_class(self, real_images, fake_images, real_labels, num_classes):
        """Calculate SSIM and PSNR per class"""
        # Denormalize images to [0, 1]
        real_images = (real_images + 1) / 2.0
        fake_images = (fake_images + 1) / 2.0

        class_ssim = {}
        class_psnr = {}

        for class_idx in range(num_classes):
            class_mask = (real_labels == class_idx)
            if class_mask.sum() == 0:
                continue

            real_class = real_images[class_mask]
            fake_class = fake_images[class_mask]

            ssim_scores = []
            psnr_scores = []

            # Calculate for each image pair in the class
            for real_img, fake_img in zip(real_class, fake_class):
                real_np = real_img.cpu().numpy().transpose(1, 2, 0)
                fake_np = fake_img.cpu().numpy().transpose(1, 2, 0)

                # SSIM
                ssim_score = ssim(real_np, fake_np, multichannel=True, data_range=1.0)
                ssim_scores.append(ssim_score)

                # PSNR
                psnr_score = psnr(real_np, fake_np, data_range=1.0)
                psnr_scores.append(psnr_score)

            class_ssim[class_idx] = np.mean(ssim_scores) if ssim_scores else 0.0
            class_psnr[class_idx] = np.mean(psnr_scores) if psnr_scores else 0.0

        return class_ssim, class_psnr

    def calculate_all_metrics(self, real_images, fake_images, real_labels, num_classes):
        """Calculate all evaluation metrics"""
        # FID Score
        real_features = self.get_inception_features(real_images)
        fake_features = self.get_inception_features(fake_images)
        fid_score = self.calculate_fid(real_features, fake_features)

        # SSIM and PSNR per class
        class_ssim, class_psnr = self.calculate_ssim_psnr_per_class(
            real_images, fake_images, real_labels, num_classes
        )

        # Overall SSIM and PSNR
        overall_ssim = np.mean(list(class_ssim.values())) if class_ssim else 0.0
        overall_psnr = np.mean(list(class_psnr.values())) if class_psnr else 0.0

        return {
            'fid': fid_score,
            'overall_ssim': overall_ssim,
            'overall_psnr': overall_psnr,
            'class_ssim': class_ssim,
            'class_psnr': class_psnr
        }

def get_memory_usage():
    if torch.cuda.is_available():
        return f"{torch.cuda.memory_allocated() / 1024**3:.2f}GB"
    return "CPU mode"

class SimpleConv2d(nn.Module):
    """Lightweight convolution block"""
    def __init__(self, in_channels, out_channels, kernel_size=3, stride=1, padding=1):
        super().__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride, padding, bias=False)
        self.norm = nn.BatchNorm2d(out_channels)
        self.act = nn.LeakyReLU(0.2, inplace=True)

    def forward(self, x):
        return self.act(self.norm(self.conv(x)))

class StyleMLP(nn.Module):
    """Simplified style mapping network"""
    def __init__(self, z_dim, style_dim, num_classes):
        super().__init__()
        self.class_embed = nn.Embedding(num_classes, z_dim)
        self.mlp = nn.Sequential(
            nn.Linear(z_dim, style_dim),
            nn.LeakyReLU(0.2),
            nn.Linear(style_dim, style_dim),
            nn.LeakyReLU(0.2),
            nn.Linear(style_dim, style_dim)
        )

    def forward(self, z, class_labels=None):
        if class_labels is not None:
            z = z + self.class_embed(class_labels)
        return self.mlp(z)

class LightweightGenerator(nn.Module):
    """Ultra-lightweight generator for Colab"""
    def __init__(self, z_dim=128, style_dim=128, channels=3, size=64, num_classes=10):
        super().__init__()
        self.size = size

        # Style mapping
        self.mapping = StyleMLP(z_dim, style_dim, num_classes)

        # Initial constant
        self.constant = nn.Parameter(torch.randn(1, 64, 4, 4))

        # Upsampling layers with correct channel progression
        self.up1 = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),
            SimpleConv2d(64, 64)  # 64->64
        )

        self.up2 = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),
            SimpleConv2d(64, 32)  # 64->32
        )

        self.up3 = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),
            SimpleConv2d(32, 16)  # 32->16
        )

        self.up4 = nn.Sequential(
            nn.Upsample(scale_factor=2, mode='nearest'),
            SimpleConv2d(16, 8)   # 16->8
        )

        # Final output
        self.to_rgb = nn.Conv2d(8, channels, 1)

        # Style injection layers - match channel dimensions
        self.style1 = nn.Linear(style_dim, 64)   # For 64 channels
        self.style2 = nn.Linear(style_dim, 64)   # For 64 channels (before up2)
        self.style3 = nn.Linear(style_dim, 32)   # For 32 channels (before up3)
        self.style4 = nn.Linear(style_dim, 16)   # For 16 channels (before up4)

    def forward(self, z, class_labels=None, truncation=1.0):
        batch_size = z.shape[0]

        # Get style
        style = self.mapping(z, class_labels)

        # Start with constant (batch_size, 64, 4, 4)
        x = self.constant.repeat(batch_size, 1, 1, 1)

        # Apply style modulation and upsampling with correct dimensions
        x = x * self.style1(style).view(batch_size, 64, 1, 1)  # 64 channels
        x = self.up1(x)  # Still 64 channels after up1

        x = x * self.style2(style).view(batch_size, 64, 1, 1)  # 64 channels
        x = self.up2(x)  # Now 32 channels after up2

        x = x * self.style3(style).view(batch_size, 32, 1, 1)  # 32 channels
        x = self.up3(x)  # Now 16 channels after up3

        x = x * self.style4(style).view(batch_size, 16, 1, 1)  # 16 channels
        x = self.up4(x)  # Now 8 channels after up4

        # Convert to RGB
        x = self.to_rgb(x)
        return torch.tanh(x)

class LightweightDiscriminator(nn.Module):
    """Ultra-lightweight discriminator for Colab"""
    def __init__(self, channels=3, size=64, num_classes=10):
        super().__init__()

        # Downsampling path
        self.down1 = SimpleConv2d(channels, 16, stride=2)      # 64->32
        self.down2 = SimpleConv2d(16, 32, stride=2)            # 32->16
        self.down3 = SimpleConv2d(32, 64, stride=2)            # 16->8
        self.down4 = SimpleConv2d(64, 128, stride=2)           # 8->4

        # Final layers
        self.final = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten()
        )

        self.real_fake = nn.Linear(128, 1)
        self.classifier = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.down1(x)
        x = self.down2(x)
        x = self.down3(x)
        x = self.down4(x)
        x = self.final(x)

        real_fake = self.real_fake(x)
        class_pred = self.classifier(x)

        return real_fake, class_pred

class ClassDataset(Dataset):
    def __init__(self, data_dir, image_size=64, max_samples_per_class=100):  # Limit samples
        self.data_dir = Path(data_dir)
        self.image_size = image_size

        self.images = []
        self.labels = []
        self.class_names = []

        print(f"Loading dataset from: {self.data_dir}")

        if not self.data_dir.exists():
            raise FileNotFoundError(f"Data directory not found: {self.data_dir}")

        for class_idx, class_dir in enumerate(sorted(self.data_dir.iterdir())):
            if class_dir.is_dir():
                self.class_names.append(class_dir.name)
                print(f"Processing class {class_idx}: {class_dir.name}")

                image_files = []
                for ext in ['*.jpg', '*.jpeg', '*.png', '*.JPG', '*.JPEG', '*.PNG']:
                    image_files.extend(list(class_dir.glob(ext)))

                # Limit samples per class to reduce memory
                image_files = image_files[:max_samples_per_class]
                print(f"  Using {len(image_files)} images (limited for memory)")

                for img_path in image_files:
                    self.images.append(img_path)
                    self.labels.append(class_idx)

        self.num_classes = len(self.class_names)

        print(f"Dataset loaded: {len(self.images)} total images, {self.num_classes} classes")

        if len(self.images) == 0:
            raise ValueError(f"No images found in {self.data_dir}")

        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        ])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.images[idx]).convert('RGB')
            return self.transform(image), self.labels[idx]
        except Exception as e:
            # Return black image if error
            dummy = Image.new('RGB', (self.image_size, self.image_size), 'black')
            return self.transform(dummy), self.labels[idx]

class MinimalTrainer:
    def __init__(self, generator, discriminator, device='cuda'):
        self.generator = generator.to(device)
        self.discriminator = discriminator.to(device)
        self.device = device

        # Lighter optimizers
        self.opt_g = optim.AdamW(generator.parameters(), lr=0.001, betas=(0.5, 0.9), weight_decay=0.01)
        self.opt_d = optim.AdamW(discriminator.parameters(), lr=0.001, betas=(0.5, 0.9), weight_decay=0.01)

        self.adversarial_loss = nn.BCEWithLogitsLoss()
        self.classification_loss = nn.CrossEntropyLoss()

        # Initialize comprehensive metrics calculator
        self.metrics_calculator = ComprehensiveMetrics(device)

        self.history = {
            'g_loss': [],
            'd_loss': [],
            'fid_scores': [],
            'overall_ssim': [],
            'overall_psnr': [],
            'class_ssim': [],  # List of dictionaries
            'class_psnr': []   # List of dictionaries
        }

    def train_step(self, real_images, real_labels):
        batch_size = real_images.shape[0]

        # Train Discriminator
        self.opt_d.zero_grad()

        # Real images
        real_pred, real_class_pred = self.discriminator(real_images)
        d_loss_real = self.adversarial_loss(real_pred, torch.ones_like(real_pred))
        d_loss_class = self.classification_loss(real_class_pred, real_labels)

        # Fake images
        z = torch.randn(batch_size, 128, device=self.device)
        with torch.no_grad():  # Detach to save memory
            fake_images = self.generator(z, real_labels)

        fake_pred, _ = self.discriminator(fake_images)
        d_loss_fake = self.adversarial_loss(fake_pred, torch.zeros_like(fake_pred))

        d_loss = d_loss_real + d_loss_fake + d_loss_class
        d_loss.backward()
        self.opt_d.step()

        # Train Generator
        self.opt_g.zero_grad()

        # Generate new fake images for generator training
        z = torch.randn(batch_size, 128, device=self.device)
        fake_images = self.generator(z, real_labels)
        fake_pred, fake_class_pred = self.discriminator(fake_images)

        g_loss_adv = self.adversarial_loss(fake_pred, torch.ones_like(fake_pred))
        g_loss_class = self.classification_loss(fake_class_pred, real_labels)
        g_loss = g_loss_adv + g_loss_class

        g_loss.backward()
        self.opt_g.step()

        # Cleanup
        del fake_images, real_pred, fake_pred
        clear_memory()

        return g_loss.item(), d_loss.item()

    def evaluate_comprehensive_metrics(self, dataloader, num_samples=100):
        """Calculate comprehensive evaluation metrics"""
        self.generator.eval()

        real_images_list = []
        fake_images_list = []
        real_labels_list = []

        with torch.no_grad():
            samples_collected = 0
            for real_images, labels in dataloader:
                if samples_collected >= num_samples:
                    break

                real_images = real_images.to(self.device)
                labels = labels.to(self.device)

                # Generate fake images
                z = torch.randn(real_images.shape[0], 128, device=self.device)
                fake_images = self.generator(z, labels)

                real_images_list.append(real_images)
                fake_images_list.append(fake_images)
                real_labels_list.append(labels)

                samples_collected += real_images.shape[0]
                clear_memory()

        # Concatenate all samples
        real_images_all = torch.cat(real_images_list, dim=0)
        fake_images_all = torch.cat(fake_images_list, dim=0)
        real_labels_all = torch.cat(real_labels_list, dim=0)

        # Calculate all metrics
        metrics = self.metrics_calculator.calculate_all_metrics(
            real_images_all, fake_images_all, real_labels_all,
            num_classes=len(set(real_labels_all.cpu().numpy()))
        )

        self.generator.train()
        clear_memory()

        return metrics

    def train(self, dataloader, num_epochs, save_dir='./checkpoints', class_names=None):
        Path(save_dir).mkdir(parents=True, exist_ok=True)

        print(f"Starting training with {get_memory_usage()}")

        for epoch in range(num_epochs):
            epoch_g_loss = 0
            epoch_d_loss = 0

            pbar = tqdm(dataloader, desc=f'Epoch {epoch+1}/{num_epochs}')

            for batch_idx, (real_images, labels) in enumerate(pbar):
                real_images = real_images.to(self.device)
                labels = labels.to(self.device)

                g_loss, d_loss = self.train_step(real_images, labels)
                epoch_g_loss += g_loss
                epoch_d_loss += d_loss

                # Update progress
                pbar.set_postfix({
                    'G': f'{g_loss:.3f}',
                    'D': f'{d_loss:.3f}',
                    'Mem': get_memory_usage()
                })

                # Clear memory every 5 batches
                if batch_idx % 5 == 0:
                    clear_memory()

            # Average losses
            epoch_g_loss /= len(dataloader)
            epoch_d_loss /= len(dataloader)

            self.history['g_loss'].append(epoch_g_loss)
            self.history['d_loss'].append(epoch_d_loss)

            # Comprehensive evaluation every 10 epochs
            if (epoch + 1) % 10 == 0:
                print(f"\nEvaluating comprehensive metrics at epoch {epoch+1}...")
                try:
                    metrics = self.evaluate_comprehensive_metrics(dataloader)

                    # Store metrics
                    self.history['fid_scores'].append(metrics['fid'])
                    self.history['overall_ssim'].append(metrics['overall_ssim'])
                    self.history['overall_psnr'].append(metrics['overall_psnr'])
                    self.history['class_ssim'].append(metrics['class_ssim'])
                    self.history['class_psnr'].append(metrics['class_psnr'])

                    # Print detailed results
                    print(f'Epoch {epoch+1}: G Loss: {epoch_g_loss:.4f}, D Loss: {epoch_d_loss:.4f}')
                    print(f'  FID Score: {metrics["fid"]:.3f}')
                    print(f'  Overall SSIM: {metrics["overall_ssim"]:.3f}')
                    print(f'  Overall PSNR: {metrics["overall_psnr"]:.3f} dB')

                    # Print per-class metrics
                    print("  Per-Class SSIM:")
                    for class_idx, ssim_val in metrics['class_ssim'].items():
                        class_name = class_names[class_idx] if class_names else f"Class {class_idx}"
                        print(f"    {class_name}: {ssim_val:.3f}")

                    print("  Per-Class PSNR:")
                    for class_idx, psnr_val in metrics['class_psnr'].items():
                        class_name = class_names[class_idx] if class_names else f"Class {class_idx}"
                        print(f"    {class_name}: {psnr_val:.3f} dB")

                except Exception as e:
                    print(f"  Error calculating metrics: {e}")
                    # Add placeholder values
                    self.history['fid_scores'].append(float('inf'))
                    self.history['overall_ssim'].append(0.0)
                    self.history['overall_psnr'].append(0.0)
                    self.history['class_ssim'].append({})
                    self.history['class_psnr'].append({})

                # Save checkpoint
                torch.save({
                    'generator': self.generator.state_dict(),
                    'discriminator': self.discriminator.state_dict(),
                    'epoch': epoch,
                    'history': self.history
                }, f'{save_dir}/checkpoint_epoch_{epoch+1}.pth')

                clear_memory()
            else:
                print(f'Epoch {epoch+1}: G Loss: {epoch_g_loss:.4f}, D Loss: {epoch_d_loss:.4f}')

    def generate_images(self, class_labels, num_images_per_class, save_dir):
        self.generator.eval()

        with torch.no_grad():
            for class_idx in class_labels:
                images = []

                # Generate in very small batches
                for i in range(0, num_images_per_class, 4):
                    batch_size = min(4, num_images_per_class - i)

                    z = torch.randn(batch_size, 128, device=self.device)
                    labels = torch.full((batch_size,), class_idx, device=self.device)

                    fake_images = self.generator(z, labels)
                    images.extend(fake_images.cpu())

                    clear_memory()

                # Save images
                class_dir = Path(save_dir) / f'class_{class_idx}'
                class_dir.mkdir(parents=True, exist_ok=True)

                for i, img in enumerate(images):
                    img_pil = transforms.ToPILImage()(img * 0.5 + 0.5)
                    img_pil.save(class_dir / f'generated_{i:04d}.png')

                print(f"Generated {len(images)} images for class {class_idx}")

        self.generator.train()

    def plot_comprehensive_metrics(self, save_path, class_names=None):
        """Plot comprehensive evaluation metrics"""
        fig, axes = plt.subplots(3, 2, figsize=(15, 12))

        # Training losses
        axes[0, 0].plot(self.history['g_loss'], label='Generator', color='blue')
        axes[0, 0].plot(self.history['d_loss'], label='Discriminator', color='orange')
        axes[0, 0].set_title('Training Losses')
        axes[0, 0].set_xlabel('Epoch')
        axes[0, 0].set_ylabel('Loss')
        axes[0, 0].legend()
        axes[0, 0].grid(True)

        # FID Score over epochs
        if self.history['fid_scores']:
            eval_epochs = list(range(10, len(self.history['fid_scores']) * 10 + 1, 10))
            axes[0, 1].plot(eval_epochs, self.history['fid_scores'], 'g-o', markersize=4)
            axes[0, 1].set_title('FID Score Over Epochs')
            axes[0, 1].set_xlabel('Epoch')
            axes[0, 1].set_ylabel('FID Score')
            axes[0, 1].grid(True)

        # Overall SSIM and PSNR
        if self.history['overall_ssim'] and self.history['overall_psnr']:
            eval_epochs = list(range(10, len(self.history['overall_ssim']) * 10 + 1, 10))
            axes[1, 0].plot(eval_epochs, self.history['overall_ssim'], 'r-o', markersize=4)
            axes[1, 0].set_title('Overall SSIM Score')
            axes[1, 0].set_xlabel('Epoch')
            axes[1, 0].set_ylabel('SSIM')
            axes[1, 0].grid(True)

            axes[1, 1].plot(eval_epochs, self.history['overall_psnr'], 'm-o', markersize=4)
            axes[1, 1].set_title('Overall PSNR Score')
            axes[1, 1].set_xlabel('Epoch')
            axes[1, 1].set_ylabel('PSNR (dB)')
            axes[1, 1].grid(True)

        # Per-class SSIM (latest epoch)
        if self.history['class_ssim']:
            latest_ssim = self.history['class_ssim'][-1]
            classes = list(latest_ssim.keys())
            ssim_values = list(latest_ssim.values())

            class_labels = [class_names[i] if class_names and i < len(class_names) else f"Class {i}"
                          for i in classes]

            axes[2, 0].bar(range(len(classes)), ssim_values, color='skyblue')
            axes[2, 0].set_title('SSIM Score per Class (Latest Epoch)')
            axes[2, 0].set_xlabel('Class')
            axes[2, 0].set_ylabel('SSIM')
            axes[2, 0].set_xticks(range(len(classes)))
            axes[2, 0].set_xticklabels(class_labels, rotation=45, ha='right')
            axes[2, 0].grid(True, alpha=0.3)

        # Per-class PSNR (latest epoch)
        if self.history['class_psnr']:
            latest_psnr = self.history['class_psnr'][-1]
            classes = list(latest_psnr.keys())
            psnr_values = list(latest_psnr.values())

            class_labels = [class_names[i] if class_names and i < len(class_names) else f"Class {i}"
                          for i in classes]

            axes[2, 1].bar(range(len(classes)), psnr_values, color='lightcoral')
            axes[2, 1].set_title('PSNR Score per Class (Latest Epoch)')
            axes[2, 1].set_xlabel('Class')
            axes[2, 1].set_ylabel('PSNR (dB)')
            axes[2, 1].set_xticks(range(len(classes)))
            axes[2, 1].set_xticklabels(class_labels, rotation=45, ha='right')
            axes[2, 1].grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.show()

def main():
    # Force CPU if CUDA out of memory
    try:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        # Test CUDA memory
        if device.type == 'cuda':
            test_tensor = torch.randn(100, 100, device=device)
            del test_tensor
            clear_memory()
    except RuntimeError:
        print("âš ï¸ CUDA memory issue detected, switching to CPU")
        device = torch.device('cpu')

    print(f"Using device: {device}")

    # Ultra-conservative configuration
    CONFIG = {
        'DATA_PATH': '/content/drive/Shareddrives/Research_ES/CNN_Swin/Original_SOD',
        'OUTPUT_PATH': '/content/drive/Shareddrives/Research_ES/CNN_Swin/Generated_Images_Light',
        'CHECKPOINT_PATH': '/content/drive/Shareddrives/Research_ES/CNN_Swin/Checkpoints_Light',
        'RESULTS_PATH': '/content/drive/Shareddrives/Research_ES/CNN_Swin/Results_Light',

        # Minimal parameters
        'IMAGE_SIZE': 64,    # Very small
        'BATCH_SIZE': 8,     # Very small batch
        'NUM_EPOCHS': 150,    # Fewer epochs
        'Z_DIM': 128,        # Minimal latent dim
        'STYLE_DIM': 128,    # Minimal style dim
        'MAX_SAMPLES': 100,  # Limit dataset size

        # Generation config
        'CLASS_GENERATION': {i: 10 for i in range(11)}  # Only 5 images per class
    }

    # Create directories
    for path in [CONFIG['OUTPUT_PATH'], CONFIG['CHECKPOINT_PATH'], CONFIG['RESULTS_PATH']]:
        Path(path).mkdir(parents=True, exist_ok=True)

    try:
        # Load minimal dataset
        print("=" * 50)
        print("LOADING MINIMAL DATASET")
        print("=" * 50)

        dataset = ClassDataset(
            CONFIG['DATA_PATH'],
            image_size=CONFIG['IMAGE_SIZE'],
            max_samples_per_class=CONFIG['MAX_SAMPLES']
        )

        dataloader = DataLoader(
            dataset,
            batch_size=CONFIG['BATCH_SIZE'],
            shuffle=True,
            num_workers=0,  # No multiprocessing
            drop_last=True
        )

        print(f"Dataset: {len(dataset)} images, {dataset.num_classes} classes")

        # Initialize lightweight models
        print("\n" + "=" * 50)
        print("INITIALIZING LIGHTWEIGHT MODELS")
        print("=" * 50)

        generator = LightweightGenerator(
            z_dim=CONFIG['Z_DIM'],
            style_dim=CONFIG['STYLE_DIM'],
            size=CONFIG['IMAGE_SIZE'],
            num_classes=dataset.num_classes
        )

        discriminator = LightweightDiscriminator(
            size=CONFIG['IMAGE_SIZE'],
            num_classes=dataset.num_classes
        )

        # Model info
        g_params = sum(p.numel() for p in generator.parameters())
        d_params = sum(p.numel() for p in discriminator.parameters())

        print(f"Generator parameters: {g_params:,}")
        print(f"Discriminator parameters: {d_params:,}")
        print(f"Total parameters: {g_params + d_params:,}")
        print(f"Memory: {get_memory_usage()}")

        # Train
        trainer = MinimalTrainer(generator, discriminator, device)

        print("\n" + "=" * 50)
        print("STARTING COMPREHENSIVE TRAINING")
        print("=" * 50)

        trainer.train(dataloader, CONFIG['NUM_EPOCHS'], CONFIG['CHECKPOINT_PATH'],
                     class_names=dataset.class_names)

        # Generate images
        print("\n" + "=" * 50)
        print("GENERATING IMAGES")
        print("=" * 50)

        for class_idx, num_images in CONFIG['CLASS_GENERATION'].items():
            if class_idx < dataset.num_classes:
                trainer.generate_images([class_idx], num_images, CONFIG['OUTPUT_PATH'])

        # Save comprehensive results
        print("\n" + "=" * 50)
        print("SAVING COMPREHENSIVE RESULTS")
        print("=" * 50)

        trainer.plot_comprehensive_metrics(
            f"{CONFIG['RESULTS_PATH']}/comprehensive_metrics.png",
            class_names=dataset.class_names
        )

        # Save detailed summary with all metrics
        summary = {
            'config': CONFIG,
            'model_params': {'generator': g_params, 'discriminator': d_params},
            'dataset_info': {
                'total_images': len(dataset),
                'num_classes': dataset.num_classes,
                'class_names': dataset.class_names
            },
            'final_metrics': {
                'fid': trainer.history['fid_scores'][-1] if trainer.history['fid_scores'] else None,
                'overall_ssim': trainer.history['overall_ssim'][-1] if trainer.history['overall_ssim'] else None,
                'overall_psnr': trainer.history['overall_psnr'][-1] if trainer.history['overall_psnr'] else None,
                'class_ssim': trainer.history['class_ssim'][-1] if trainer.history['class_ssim'] else {},
                'class_psnr': trainer.history['class_psnr'][-1] if trainer.history['class_psnr'] else {},
                'generator_loss': trainer.history['g_loss'][-1],
                'discriminator_loss': trainer.history['d_loss'][-1]
            },
            'training_history': trainer.history
        }

        with open(f"{CONFIG['RESULTS_PATH']}/comprehensive_summary.json", 'w') as f:
            json.dump(summary, f, indent=2)

        print("âœ… COMPREHENSIVE TRAINING COMPLETED!")
        print(f"ðŸ“ Generated Images: {CONFIG['OUTPUT_PATH']}")
        print(f"ðŸ’¾ Checkpoints: {CONFIG['CHECKPOINT_PATH']}")
        print(f"ðŸ“Š Comprehensive Results: {CONFIG['RESULTS_PATH']}")

        # Print final comprehensive metrics
        if trainer.history['fid_scores']:
            final_metrics = summary['final_metrics']
            print(f"\nðŸ“Š FINAL COMPREHENSIVE METRICS:")
            print(f"   FID Score: {final_metrics['fid']:.3f}")
            print(f"   Overall SSIM: {final_metrics['overall_ssim']:.3f}")
            print(f"   Overall PSNR: {final_metrics['overall_psnr']:.3f} dB")

            print(f"\nðŸ“‹ PER-CLASS METRICS:")
            for class_idx, class_name in enumerate(dataset.class_names):
                if class_idx in final_metrics['class_ssim']:
                    ssim_val = final_metrics['class_ssim'][class_idx]
                    psnr_val = final_metrics['class_psnr'][class_idx]
                    print(f"   {class_name}: SSIM={ssim_val:.3f}, PSNR={psnr_val:.3f} dB")

    except RuntimeError as e:
        if "out of memory" in str(e).lower():
            print("\nâŒ STILL OUT OF MEMORY!")
            print("Emergency solutions:")
            print("1. Restart Colab runtime completely")
            print("2. Use CPU-only: device = torch.device('cpu')")
            print("3. Reduce BATCH_SIZE to 4")
            print("4. Reduce IMAGE_SIZE to 32")
            print("5. Reduce MAX_SAMPLES to 50")
        else:
            raise e

    except Exception as e:
        print(f"âŒ Error: {e}")
        raise e

    finally:
        clear_memory()

if __name__ == "__main__":
    main()
```
